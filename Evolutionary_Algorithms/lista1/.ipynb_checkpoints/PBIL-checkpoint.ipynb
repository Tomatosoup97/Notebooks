```python
import time
from functools import reduce, partial

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python

# d - długość chromosomu
# N - wielkość populacji
# t1 - współczynnik uczenia
# t2 - prawdopodobieństwo mutacji
# t3 - współczynnik zaburzenia podczas mutacji

d = 100
omega = [0, 1]
T = 1500
history = []
prob_v_history = []


def random_population(prob_vector, N):
    population = [
        np.random.choice(omega, (N, 1), p=[1-prob, prob])
        for prob in prob_vector
    ]
    return reduce(
        lambda pop, x: np.concatenate((pop, x), axis=1),
        population[1:],
        population[0])


def initial_prob_vector(d):
    return np.full(d, 0.5)


def best_individual(F, population):
    vals = F(population)
    return population[np.argmax(vals)]


def pbil(F, T, N, t1, t2, t3):
    prob_vector = initial_prob_vector(d)
    population = random_population(prob_vector, N)
    result = F(population)
    
    for i in range(T):
        if not i % (T/10):
            print('.', end='')
        x = best_individual(F, population)

        for k in range(d):
            prob_vector[k] = prob_vector[k] * (1-t1) + x[k] * t1

            if np.random.uniform() < t2:
                prob_vector[k] = prob_vector[k] * (1-t3) + np.random.choice([0, 1]) * t3

        prob_v_history.append(np.sum(prob_vector))
        population = random_population(prob_vector, N)
        result = F(population)
        history.append(result)
        
    print('Done')
```

```python
def run_pbil(F, T, N, t1, t2, t3):
    global history, prob_v_history
    history = []
    prob_v_history = []

    ts = time.time()
    pbil(F, T, N, t1, t2, t3)
    te = time.time()
    print(te - ts)

    
def random_rgb():
    return tuple(np.random.uniform() for i in range(3))
```

```python
def one_max(population):
    return np.sum(population, axis=1)


def deceptive_one_max(population):
    sums = np.sum(population, axis=1)
    for i, s in enumerate(sums):
        if s == 0:
            sums[i] = population.shape[1] + 1
    return sums


def k_deceptive_one_max(k, population):
    pops = np.hsplit(population, k)
    sums = np.array(list(map(deceptive_one_max, pops))).T
    return np.sum(population, axis=1)
```

```python
# Plot all solutions as points

def plot_all_points(history):
    plt.figure(figsize=(10, 6), dpi=80)
    #plt.figure()
    plt.xticks(np.arange(0, T+1, int(T/10)))
    plt.title('All samples - points')

    for r in np.array(history).T:
        plt.plot(range(T), r, 'ro', color=random_rgb())

    plt.show()

def plot_all_lines(history):
    plt.figure(figsize=(10, 6), dpi=80)
    #plt.figure()
    plt.xticks(np.arange(0, T+1, int(T/10)))
    plt.title('All samples - lines')

    for r in np.array(history).T:
        plt.plot(range(T), r, color=random_rgb())

    plt.show()

def plot_max_mean_min(history):
    plt.figure(figsize=(10, 6), dpi=80)
    #plt.figure()
    plt.xticks(np.arange(0, T+1, int(T/10)))

    xs = np.array(range(T))

    plt.title('max, mean, min')
    plt.plot(xs, list(map(np.max, history)))
    plt.plot(xs, list(map(np.mean, history)))
    plt.plot(xs, list(map(np.min, history)))

    plt.show()

def plot_all(history):
    plot_max_mean_min(history)
    plot_all_points(history)
    plot_all_lines(history)
    
```

```python
# One-max solution

t1 = 0.01
t2 = 0.05
t3 = 0.01
N = 125
F = one_max

run_pbil(F, T, N, t1, t2, t3)
plot_all(history)
```

```python
plt.figure()
plt.plot(range(T), prob_v_history)
plt.show()
```
