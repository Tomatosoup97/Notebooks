# Task 2

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import math


def delta(a, b, c):
    d = b*b - 4.0*a*c
    if d < 0: print("delta can't be < 0 !")
    return d


quadratic_f = lambda a, b, c: lambda x: a * x**2 + b * x + c


def output(f, x1, x2):
    print(f'x1={x1}, f(x1)={f(x1)}')
    print(f'x2={x2}, f(x2)={f(x2)}')
    print('---------------------- \n')


def calc_using_delta(a, b, c):
    d = delta(a, b ,c)
    
    f = quadratic_f(a, b, c)
    
    x1 = (-b - math.sqrt(d)) / (2*a)
    x2 = (-b + math.sqrt(d)) / (2*a)

    output(f, x1, x2)
    return f, x1, x2


def calc_using_viete(a, b, c):
    d = delta(a, b ,c)

    x1 = (-b + math.sqrt(d) * (-np.sign(b))) / (2.0*a)
    x2 = c / (a * x1)

    f = quadratic_f(a, b, c)
    output(f, x1, x2)
    return f, x1, x2


print('\n"School" method: \n\n')

a1, b1, c1 = 1e-5, 10 ** 12, 1.0
a2, b2, c2 = 1.0, 10 ** 7, 1e-3
a3, b3, c3 = 3e-5, 5**7, 5e-3

calc_using_delta(a1, b1, c1)

calc_using_delta(a2, b2, c2)

calc_using_delta(a3, b3, c3)

print('\nViete formula\n\n')

calc_using_viete(a1, b1, c1)

calc_using_viete(a3, b3, c3)
```

# Task 3

```python

def bad_calc(r, q):
    k = math.sqrt(q**3 + r**2)
    a = (r + k)**(1./3.)
    if r-k < 0:
        b = -(-(r - k))**(1./3.)
    else:
        b = (r - k)**(1./3.)
    return a + b


def superior_calc(r, q):
    cube_f = lambda r, q: (r + math.sqrt(q**3 + r**2))**(1/3)
    k = cube_f(r, q)
    
    return (2*r * k**2) / (k**4 + k**2 * q + q**2)


q = 1
r = 10**11

print(f'1st version: {bad_calc(q, r)}')
print(f'2nd version: {superior_calc(q, r)}')


q = 1e-4
r = 10**8

print(f'\n1st version: {bad_calc(q, r)}')
print(f'2nd version: {superior_calc(q, r)}')
```

# Task 5

```python
def output_results(f, iterable):
    results = list(map(f, iterable))

    plt.plot(list(iterable), results, 'k')
    plt.grid(True)
    plt.show()


f_a = lambda x: x**2 / abs(1-x**2)

output_results(f_a, list(range(-40, -1)) + list(range(2, 40)))


f_b = lambda x: abs(x / math.tan(x) - 1)

output_results(f_b, [2, 3, math.pi - 0.1, math.pi, math.pi + 0.1, 4])


f_c = lambda x: abs(-3*x*math.tan(3*x))

output_results(f_c, range(-20, 20))


f_d_aux = lambda x: math.sqrt(x**2 + 2017)


f_d = lambda x: abs( (x/f_d_aux(x) - 1) * x / (f_d_aux(x) - x) )

output_results(f_d, range(-100, 100))
output_results(f_d, list(map(lambda n: n/1000, range(-100, 100))))

print(list(map(f_d, [-0.5, -0.2, -0.1, 0, 0.1, 0.2, 0.5])))
```
